```python
import seaborn as sns
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.isnull().sum()
```

```python
titanic.dtypes
```

```python
titanic = titanic.drop('deck',axis=1)
```

```python
titanic.isnull().sum()
```

```python
import numpy as np
```

```python
titanic['age'] = titanic['age'].replace(np.NaN,titanic['age'].mean())
```

```python
titanic.isnull().sum()
```

```python
titanic = titanic.dropna()
```

```python
titanic.isnull().sum()
```

```python
sns.boxplot(x='sex',y='age',data=titanic, hue='survived')
```

```python
sns.histplot(x='fare',data=titanic,stat='frequency')
```

```python
sns.violinplot(x='sex',y='age',data=titanic,hue='survived')
```

```python
sns.swarmplot(x='sex',y='age',data=titanic,hue='survived')
```
