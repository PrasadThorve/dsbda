```python
import pandas as pd
```

```python
DataFrame = pd.read_csv("dirtydata.csv")
```

```python
#print(df)
df = DataFrame.copy()
```

```python
#isnull() function is used to find missing values
df.isnull()
```

```python
df.isnull().sum()
```

```python
# dropna() function is used to remove missing values
df.dropna()
```

dropna() function delete entire record

```python
# fillna() function is used to fill missing values
# df.fillna(130,inplace=True)
```

```python
# We can replace null values using mean, median and mode of respective column.

x = df["Calories"].mean()
print(x)
df["Calories"].fillna(x,inplace=True)
df
```

```python
#Reshaping data
df['Gender']=['M','M','F','F','M','M','F','M','F','F','M','M','F','M','F','F','M','M','F','M','F','F','M','M','F','M','F','F','M','M','F','M']
df
```

```python
#Transforming data using function or Mapping
df['Gender']=df['Gender'].map({'M':1,'F':0}).astype(float)
df
```

```python
# Discovering duplicate

df.duplicated()
```

```python
# data wrangling by removing duplicates we can remove duplicates by using drop_duplicates() fuction
df.drop_duplicates(inplace=True)
```

```python
#describe() is used to write description of the data
df.describe()
```
